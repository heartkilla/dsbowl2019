```python
import pandas as pd
import numpy as np
```

```python
train_dataset = pd.read_csv("data-science-bowl-2019/train.csv")
```

```python
train_labels = pd.read_csv("data-science-bowl-2019/train_labels.csv")
```

```python
installations_in_labels = train_labels["installation_id"].unique()
del train_labels
```

```python
train_dataset = train_dataset[train_dataset["installation_id"].isin(installations_in_labels)]
```

```python
train_dataset["timestamp"] = pd.to_datetime(train_dataset["timestamp"], format = "%Y-%m-%d %H:%M:%S")
train_dataset = train_dataset.sort_values(["installation_id", "timestamp"])
```

```python
pd.set_option('max_row', 1000)
train_dataset[train_dataset["type"] == "Assessment"].head(n=1000)
```

```python
groups = train_dataset[:1000000].groupby("installation_id")
```

```python
#TODO fix issue with incorrect order of events in assessment list
#TODO decide what to do with taken but not completed assessments
entries_list = []
current_group = 0
for group in groups:
    print("Group {} out of {} is under processing".format(current_group, len(groups)))
    titles = []
    event_count = 0
    event_codes = []
    event_ids = []
    prev_title = ""
    prev_code = ""
    #prev_session_event_count = 0
    prev_assessment_event_count = 0
    prev_assessment_name = ""
    #magma_events(and others as well)= 
    #prev_Assessment_time(start&end)
    #prev_activity_event_count
    current_game_session = []
    for index, row in group[1].iterrows():
        if row["type"] == "Assessment":
            if current_game_session == row["game_session"]:
                prev_assessment_event_count += 1
            else:
                #prev_assessment_completion
                # create a new entry for the train dataset 
                entry_row = {}
                entry_row["game_session"] = row["game_session"]
                entry_row["installation_id"] = row["installation_id"]
                entry_row["event_count"] =  event_count
                entry_row["prev_assessment_event_count"] = prev_assessment_event_count
                entry_row["titles_number"] = len(set(titles))
                entry_row["prev_code"] = prev_code
                entry_row["event_codes_number"] = len(set(event_codes))
                entry_row["event_ids_number"] = len(set(event_ids)) 
                entry_row["prev_title"] = prev_title
                entry_row["prev_assessment_name"] = prev_assessment_name
                #entry_row["most_frequent_event_id"]
                #entry_row["most_frequent_title"] 
                #entry_row["most_frequent_event_code"]
                #entry_row["not_attentive_code_count"]
                entries_list.append(entry_row)
                
                # update service variables
                current_game_session = row["game_session"]
                prev_assessment_event_count = 1
                prev_assessment_name = row["title"]
        event_count += 1
        titles.append(row["title"])
        prev_code = row["event_code"]
        event_codes.append(row["event_code"])
        event_ids.append(row["event_id"])
        prev_title = row["title"]
        #most_frequent_title
    current_group += 1
train_df = pd.DataFrame(entries_list)
print(train_df)
```
