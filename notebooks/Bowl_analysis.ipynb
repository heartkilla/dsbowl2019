## Distribution of events throught the year

As can be seen above events start at 2019-07-23 02:12:17.279 and last event happened at 2019-10-22 17:41:45.983. Let's see the events distribution during they year

```python
df["date"] = df["timestamp"].dt.date
df["hour"] = df["timestamp"].dt.hour
df["weekday"] = df["timestamp"].dt.weekday
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
df.groupby(["date"])["event_id"].count().plot()
fig.autofmt_xdate()
ax.set_ylabel('Number of events per day')
fig.autofmt_xdate()
fig.savefig('events_per_day.png', bbox_inches='tight')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
df.groupby(["date"])["game_session"].nunique().plot()
fig.autofmt_xdate()
ax.set_ylabel('Number of sessions per day')
fig.autofmt_xdate()
fig.savefig('sessions_per_day.png', bbox_inches='tight')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
df.groupby(["date"])["installation_id"].nunique().plot()
fig.autofmt_xdate()
ax.set_ylabel('Number of unique users per day')
fig.savefig('users_per_day.png', bbox_inches='tight')
```

In the next chart I select *max()* since it allows to get the last event in the session

```python
fig = plt.figure()
ax = fig.add_subplot(111)
last_session_events = df.groupby(["game_session"])["game_time", "date", "hour", "weekday"].max()
#miliseconds to seconds
last_session_events["game_time"] /= 1000
last_session_events.groupby("date")["game_time"].mean().plot()
fig.autofmt_xdate()
ax.set_ylabel('Average session gaming time per day in seconds')
fig.savefig('gaming_time_per_day.png', bbox_inches='tight')
```

Let's see an overall distribution of the *game_time*:

```python
last_session_events.describe()
```

As we can see from this statistics, more than 50% of game sessions consist of only one event. For that reason, let's split the game sessions which contain only one event and which contain many events so playing time is higher than 0:

```python
gaming_0_time = last_session_events[last_session_events["game_time"] == 0]
gaming_not_0_time = last_session_events[last_session_events["game_time"] > 0]
```

Let's see the distribution of these gaming times, how many 0 session are there per day?

```python
gaming_0_time.groupby("date")["game_time"].count().plot()
fig.autofmt_xdate()
ax.set_ylabel('0 time sessions per day')
fig.savefig('0_sessions_per_day.png', bbox_inches='tight')
```

How many not 0 sessions are there per day?

```python
gaming_not_0_time.groupby("date")["game_time"].count().plot()
fig.autofmt_xdate()
ax.set_ylabel('not 0 sessions per day')
fig.savefig('not_0_sessions_per_day.png', bbox_inches='tight')
```

what's an average playing time is there per day?

```python
gaming_not_0_time.groupby("date")["game_time"].mean().plot()
fig.autofmt_xdate()
ax.set_ylabel('average gaming time per day')
fig.savefig('average_gaming_time_per_day.png', bbox_inches='tight')
```

Let's see the distribution of eventd during the day:

```python
df.groupby("hour")["event_id"].count().plot()
ax.set_ylabel('Number of events per hour')
fig.autofmt_xdate()
fig.savefig('sessions_per_hour.png', bbox_inches='tight')
```

We can see that sessions usually take place from 10 AM till 2PM. 

Let's see the distribution of sessions starts and ends:

```python
game_session_grouped_events = df.sort_values(["game_session", "date"]).groupby("game_session")["date", "hour"]
game_session_last_event_time = game_session_grouped_events.last()
game_session_first_event_time = game_session_grouped_events.first()
```

```python
game_session_first_event_time.groupby("hour").count().plot()
ax.set_ylabel('Number started sessions per hour')
fig.autofmt_xdate()
fig.savefig('started_sessions_per_hour.png', bbox_inches='tight')
```

```python
game_session_last_event_time.groupby("hour").count().plot()
ax.set_ylabel('Number finished sessions per hour')
fig.autofmt_xdate()
fig.savefig('finished_sessions_per_hour.png', bbox_inches='tight')
```

## Distirubution of events according to the day of the week

```python
# all events
events_per_week_day = df.groupby("weekday")["event_id"].count()
```

```python
# events from session with more than one event(long session)
game_sessions_count = df.groupby("game_session")["event_id"].count().reset_index()
game_sessions_count.columns = ["game_session", "event_count"]
long_game_sessions = game_sessions_count[game_sessions_count["event_count"] > 1]["game_session"]
events_from_long_sessions_mask = df["game_session"].isin(long_game_sessions)
events_from_long_sessions = df[events_from_long_sessions_mask][["weekday", "event_id", 
                                                                "game_session", "installation_id"]]
```

```python
short_game_sessions = game_sessions_count[game_sessions_count["event_count"] == 1]["game_session"]
events_from_short_sessions_mask = df["game_session"].isin(short_game_sessions)
events_from_short_sessions = df[events_from_short_sessions_mask][["weekday", "event_id", 
                                                                  "game_session", "installation_id"]]
```

```python
events_per_week_day
```

```python
long_sessions_events_per_week_day = events_from_long_sessions.groupby("weekday")["event_id"].count()
short_sessions_events_per_week_day = events_from_short_sessions.groupby("weekday")["event_id"].count()
```

```python
long_sessions_events_per_week_day
```

```python
short_sessions_events_per_week_day
```

Let's compare the number of assignments, and overall number of events per day

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(events_per_week_day, color="r")
ax2 = ax.twinx() 
ax2.plot(df[df["type"] == "Assessment"].groupby("weekday")["event_id"].count(), color="b")
ax.set_ylabel('Overall number of events')
ax2.set_ylabel('Number of assessments')
week_days_list = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'] 
ax.set_xticks(range(len(week_days_list)))
ax.set_xticklabels(week_days_list)
fig.autofmt_xdate()
```

We can see that patterns are different, highest number of events is happening during the Friday, but number of Assessments is the highest during Saturday

Let's see how many unique users are participating in the game per day:

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(events_per_week_day, color="r")
ax2 = ax.twinx() 
ax2.plot(df.groupby("weekday")["installation_id"].nunique(), color="b")
ax.set_ylabel('Overall number of events')
ax2.set_ylabel('Number of unique users')
week_days_list = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'] 
ax.set_xticks(range(len(week_days_list)))
ax.set_xticklabels(week_days_list)
fig.autofmt_xdate()
```

From the chart above we can see that number of users is also highest during the Friday and drops during the weekend. Probably, it means that during the weekend a small amount of users submit a lot of assignments

Let's see if there are some users who login only on specific day:

```python
number_of_playing_days_for_user = df.groupby("installation_id")["weekday"].nunique().reset_index()
single_day_users = number_of_playing_days_for_user[number_of_playing_days_for_user["weekday"] == 1]["installation_id"]
single_day_users_mask = df["installation_id"].isin(single_day_users)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(df.groupby("weekday")["installation_id"].nunique(), color="b")
ax2 = ax.twinx() 
ax2.plot(df[single_day_users_mask].groupby("weekday")["installation_id"].nunique(), color="y")
ax.set_ylabel('Number of unique users')
ax2.set_ylabel('Number of one day users')
week_days_list = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'] 
ax.set_xticks(range(len(week_days_list)))
ax.set_xticklabels(week_days_list)
fig.autofmt_xdate()
```

From this chart we can notice that almost 30% of users who are logged in during the day are one day users

## Most popular titles

```python
df.groupby(["title", "type"])["event_id"].count().sort_values()
```

```python
df.groupby(["title"])["game_session"].nunique().sort_values()
```

Session for each title:

```python
title_process = df.groupby(["installation_id","game_session"])["timestamp", 
                                               "title", "type"].agg({"timestamp":["min", "max"], 
                                                                     "title":"first", "type":"last"}).reset_index()
title_process.columns = ["installation_id", "game_session", "start", "end", "title", "type"]
users_with_assessment = title_process[title_process["type"] == "Assessment"]["installation_id"] 
title_process = title_process[title_process["installation_id"].isin(users_with_assessment)]
title_process.loc[title_process["type"] == "Clip", "end"] = np.NaN
title_process = title_process.sort_values(["installation_id", "start"])
title_process.to_csv("data-science-bowl-2019/titlesprocess.csv", index=None)
```

```python
title_0_process = df.groupby(["installation_id","game_session"])["timestamp", 
                                               "title", "type"].agg({"timestamp":["min", "max"], 
                                                                     "title":"first", "type":"last"}).reset_index()
title_0_process.columns = ["installation_id", "game_session", "start", "end", "title", "type"]
title_0_process = title_0_process[~title_0_process["installation_id"].isin(users_with_assessment)]
title_0_process.loc[title_0_process["type"] == "Clip", "end"] = np.NaN
title_0_process = title_0_process.sort_values(["installation_id", "start"])
title_0_process.to_csv("data-science-bowl-2019/titles0process.csv", index=None)
```
